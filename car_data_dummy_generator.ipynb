```python
import pandas as pd
import numpy as np
import random

# Seed for reproducibility
np.random.seed(42)

# Sample data
makes = ['Toyota', 'Honda', 'Ford', 'Tesla', 'Chevrolet']
models = ['Model A', 'Model B', 'Model C', 'Model D', 'Model E']
fuel_types = ['Petrol', 'Diesel', 'Electric', 'Hybrid']
colors = ['White', 'Black', 'Red', 'Blue', 'Silver']
is_imported_options = [True, False]

# Generating DataFrame
df = pd.DataFrame({
    'Car ID': np.arange(1, 1001),
    'Make': np.random.choice(makes, 1000),
    'Model': np.random.choice(models, 1000),
    'Year': np.random.randint(1990, 2023, size=1000),
    'Engine Size (L)': np.round(np.random.uniform(1.0, 5.0, size=1000), 2),
    'Fuel Type': np.random.choice(fuel_types, 1000),
    'Mileage (km)': np.random.randint(0, 300000, size=1000),
    'Price': np.random.randint(5000, 80000, size=1000),
    'Color': np.random.choice(colors, 1000),
    'Is Imported': np.random.choice(is_imported_options, 1000)
})

# Display the first few rows of the DataFrame
print(df.head())
```
